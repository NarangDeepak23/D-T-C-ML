```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data
```

```python
#  target_names': array(['setosa', 'versicolor', 'virginica'], dtype='<U10')
# These are Targets Name
```

```python
data.data
```

```python
data.target
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
df.info()
```

```python
x=df.iloc[:,:-1]
y=data.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier() #Post Pruning
```

```python
dtc.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtc,filled=True)
```

```python
# Gini Index ranges from 0 to 0.5
```

```python
#  Now use hyper parameters like max _depth
dtc1=DecisionTreeClassifier(max_depth=2) #Post Pruning
dtc1.fit(x_train,y_train)

plt.figure(figsize=(15,10))
tree.plot_tree(dtc1,filled=True)
```

```python
y_pred=dtc1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# DTC using Prepruning

```python
# go to sklearn.tree library and check other hyper parameters of DTC 
```

```python
# pre-pruning
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    'ccp_alpha':[1,2,3,4,5,6,7]
}
```

```python
from sklearn.model_selection import GridSearchCV
# GridSearchCV helps to work model for all possible values of hyper parameters defined above.
```

```python
dtc2=DecisionTreeClassifier(max_depth=2)
```

```python
cv=GridSearchCV(dtc2,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred1=cv.predict(x_test)
```

```python
y_pred1
```

```python
# here seems like it is overfitted. so lets try again
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    #'ccp_alpha':[1,2,3,4,5,6,7]
}
dtc3=DecisionTreeClassifier()
cv1=GridSearchCV(dtc3,param_grid=parameter,cv=5,scoring='accuracy')
cv1.fit(x_train,y_train)
cv1.best_params_
```

```python
y_pred2=cv1.predict(x_test)
```

```python
y_pred2
```

```python
accuracy_score(y_test,y_pred2)
```

```python
accuracy_score(y_pred2,y_test)
```
